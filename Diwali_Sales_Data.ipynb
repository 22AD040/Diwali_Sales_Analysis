***EXPLORATORY DATA ANALYSIS***

***GENDER & IT'S COUNT***

```python
import seaborn as sns
import matplotlib.pyplot as plt
custom_colors = ['#FF9999', '#66B3FF']

ax = sns.countplot(x='Gender', data=df, palette=custom_colors)


for bars in ax.containers:
    ax.bar_label(bars)

# Show the plot
plt.show()
```

***GENDER & TOTAL AMOUNT***

```python
import seaborn as sns
import matplotlib.pyplot as plt

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.barplot(x='Gender', y='Amount', data=sales_gen, palette='pastel')  # You can also use 'Set2', 'husl', 'muted', etc.

plt.title('Total Amount by Gender')
plt.show()
```

***From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men***

***AGE***

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

***TOTAL AMOUNT VS AGE GROUP***

```python
import seaborn as sns
import matplotlib.pyplot as plt


sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.barplot(x='Age Group', y='Amount', data=sales_age, palette='viridis')  # You can try others like 'Set2', 'coolwarm', 'pastel', etc.

plt.title('Total Sales Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

***From above graphs we can see that most of the buyers are of age group between 26-35 yrs female***

***Total number of orders from top 10 states***

```python
import seaborn as sns
import matplotlib.pyplot as plt


sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(data=sales_state, x='State', y='Orders', palette='Set2')

plt.xticks(rotation=45)


plt.show()
```

***Total amount/sales from top 10 states***

```python
import seaborn as sns
import matplotlib.pyplot as plt

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)


sns.set(rc={'figure.figsize':(15,5)})


colors = sns.color_palette("husl", len(sales_state))

sns.barplot(data=sales_state, x='State', y='Amount', palette=colors)

plt.title('Top 10 States by Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

***Occupation***

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(rc={'figure.figsize': (20, 5)})

ax = sns.countplot(data=df, x='Occupation', palette='Set3')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.set(rc={'figure.figsize': (20, 5)})


sns.barplot(data=sales_state, x='Occupation', y='Amount', palette='Set2')

plt.xticks(rotation=45)

plt.show()
```

***Product Category***

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.set(rc={'figure.figsize':(20, 5)})


ax = sns.countplot(data=df, x='Product_Category', palette='Set2')


for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)


sns.set(rc={'figure.figsize':(20,5)})


sns.barplot(data=sales_state, x='Product_Category', y='Amount', palette='Set2')  # You can try 'Set1', 'Set2', 'Set3', 'pastel', 'husl', etc.

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

***From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category***

```python
import seaborn as sns
import matplotlib.pyplot as plt


sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)


sns.set(rc={'figure.figsize':(20, 5)})


sns.barplot(data=sales_state, x='Product_ID', y='Orders', palette='Set2')  # You can try 'Set1', 'Set2', 'Set3', 'pastel', 'deep', etc.


plt.title('Top 10 Products by Orders')
plt.xlabel('Product ID')
plt.ylabel('Total Orders')

plt.show()
```

 ***top 10 most sold products (same thing as above)***

```python
import matplotlib.pyplot as plt


colors = ['red', 'green', 'blue', 'orange', 'purple', 'cyan', 'magenta', 'brown', 'pink', 'gray']


fig1, ax1 = plt.subplots(figsize=(12, 7))
top_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)


top_products.plot(kind='bar', color=colors, ax=ax1)


ax1.set_title('Top 10 Products by Orders')
ax1.set_xlabel('Product ID')
ax1.set_ylabel('Total Orders')

plt.tight_layout()
plt.show()
```

***CONCLUSION :***

***Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category***
